Text **Cleaning**

```python
import re
```

```python
def clean_text(text):
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    # Remove non-word characters (punctuation, special characters, etc.)
    text = re.sub(r'\W', ' ', text)
    # Convert to lowercase
    text = text.lower()
    return text.strip()
```

Relation Extraction

```python
import spacy
from spacy.tokens import Span
```

```python
# Define the relation extraction function
def extract_relations(doc):
    relations = []
    for sent in doc.sents:
        spans = list(sent.ents)
        for i, span1 in enumerate(spans[:-1]):
            for span2 in spans[i + 1:]:
                relation = (span1.text, span1.label_, span2.text, span2.label_)
                relations.append(relation)
    return relations
```

```python
# Apply the relation extraction function
df['relations'] = df['text'].apply(lambda x: extract_relations(nlp(x)))
```

Building the Knowledge Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Initialize the graph
G = nx.DiGraph()
```

```python
# Add nodes and edges to the graph
for _, row in df.iterrows():
    for relation in row['relations']:
        entity1, label1, entity2, label2 = relation
        G.add_node(entity1, label=label1)
        G.add_node(entity2, label=label2)
        G.add_edge(entity1, entity2, label=f"{label1}-{label2}")
```

```python
# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.5)  # Positioning the nodes
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", edge_color="gray")
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.title("Hotel Reviews Knowledge Graph")
plt.show()
```

simplify the Knowledge Graph

```python
# Filter relations to include only specific types
def filter_relations(relations):
    filtered = []
    for relation in relations:
        entity1, label1, entity2, label2 = relation
        if label1 in ['ORG', 'GPE'] and label2 in ['ORG', 'GPE']:
            filtered.append(relation)
    return filtered

df['filtered_relations'] = df['relations'].apply(filter_relations)
```

```python
# Create the simplified graph
G_filtered = nx.DiGraph()
```

```python
for _, row in df.iterrows():
    for relation in row['filtered_relations']:
        entity1, label1, entity2, label2 = relation
        G_filtered.add_node(entity1, label=label1)
        G_filtered.add_node(entity2, label=label2)
        G_filtered.add_edge(entity1, entity2, label=f"{label1}-{label2}")
```

```python
# Draw the simplified graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G_filtered, k=0.5)
nx.draw(G_filtered, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", edge_color="gray")
edge_labels = nx.get_edge_attributes(G_filtered, 'label')
nx.draw_networkx_edge_labels(G_filtered, pos, edge_labels=edge_labels, font_color='red')
plt.title("Simplified Hotel Reviews Knowledge Graph")
plt.show()
```

```python
from networkx.algorithms import community

# Detect communities
communities = community.greedy_modularity_communities(G_filtered)
community_map = {}
for i, comm in enumerate(communities):
    for node in comm:
        community_map[node] = i

# Draw the graph with communities
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G_filtered, k=0.5)
colors = [community_map[node] for node in G_filtered]
nx.draw(G_filtered, pos, with_labels=True, node_size=3000, cmap=plt.cm.rainbow, node_color=colors, font_size=10, font_weight="bold", edge_color="gray")
nx.draw_networkx_edge_labels(G_filtered, pos, edge_labels=edge_labels, font_color='red')
plt.title("Hotel Reviews Knowledge Graph with Communities")
plt.show()
```

**Analysis** **between** **complain** **or** **Noncomplain**

```python
import pandas as pd
import re
import spacy
import networkx as nx
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data_raja_babu (500).csv")

# Preprocessing: clean and fill missing values
df = df.dropna(subset=['text'])
df.fillna({
    'complain': 0,
    'noncomplain': 0,
    'Complain Texts': '',
    'Positive Feedback': 0,
    'Negative Feedback': 0,
    'joy': 0,
    'sadness': 0,
    'fear': 0,
    'anger': 0,
    'anticipation': 0,
    'surprise': 0,
    'disgust': 0,
    'trust': 0,
    'ridicule': 0,
    'other': 0,
    'Remarks': ''
}, inplace=True)

# Clean text function
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'\W', ' ', text)
    text = text.lower()
    return text.strip()

df['text'] = df['text'].apply(clean_text)
```

```python
# Load the pre-trained NLP model
nlp = spacy.load('en_core_web_sm')

def extract_entities(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

def extract_relations(text):
    doc = nlp(text)
    relations = []
    for sent in doc.sents:
        spans = list(sent.ents)
        for i, span1 in enumerate(spans[:-1]):
            for span2 in spans[i + 1:]:
                relation = (span1.text, span1.label_, span2.text, span2.label_)
                relations.append(relation)
    return relations

df['entities'] = df['text'].apply(extract_entities)
df['relations'] = df['text'].apply(extract_relations)
```

```python
def build_knowledge_graph(dataframe):
    G = nx.DiGraph()
    for _, row in dataframe.iterrows():
        for relation in row['relations']:
            entity1, label1, entity2, label2 = relation
            G.add_node(entity1, label=label1)
            G.add_node(entity2, label=label2)
            G.add_edge(entity1, entity2, label=f"{label1}-{label2}")
    return G

# Separate complaints and non-complaints
complaints_df = df[df['complain'] == 1]
noncomplaints_df = df[df['noncomplain'] == 1]

# Build graphs for complaints and non-complaints
G_complaints = build_knowledge_graph(complaints_df)
G_noncomplaints = build_knowledge_graph(noncomplaints_df)

# Draw function for visualization
def draw_graph(G, title):
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, k=0.5)
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", edge_color="gray")
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
    plt.title(title)
    plt.show()

# Draw the graphs
draw_graph(G_complaints, "Knowledge Graph for Complaints")
draw_graph(G_noncomplaints, "Knowledge Graph for Non-Complaints")
```

```python
def graph_analysis(G):
    print("Number of nodes:", G.number_of_nodes())
    print("Number of edges:", G.number_of_edges())
    degree_centrality = nx.degree_centrality(G)
    sorted_degree = sorted(degree_centrality.items(), key=lambda item: item[1], reverse=True)
    print("Top 5 nodes by degree centrality:")
    for node, centrality in sorted_degree[:5]:
        print(f"{node}: {centrality}")

# Analyze complaints graph
print("Complaints Graph Analysis")
graph_analysis(G_complaints)

# Analyze non-complaints graph
print("Non-Complaints Graph Analysis")
graph_analysis(G_noncomplaints)
```

```python
def compare_graphs(G1, G2):
    entities_G1 = set(G1.nodes())
    entities_G2 = set(G2.nodes())

    common_entities = entities_G1.intersection(entities_G2)
    unique_to_G1 = entities_G1 - entities_G2
    unique_to_G2 = entities_G2 - entities_G1

    print("Common entities:", common_entities)
    print("Unique to complaints:", unique_to_G1)
    print("Unique to non-complaints:", unique_to_G2)

    edges_G1 = set(G1.edges())
    edges_G2 = set(G2.edges())

    common_edges = edges_G1.intersection(edges_G2)
    unique_edges_to_G1 = edges_G1 - edges_G2
    unique_edges_to_G2 = edges_G2 - edges_G1

    print("Common edges:", common_edges)
    print("Unique edges to complaints:", unique_edges_to_G1)
    print("Unique edges to non-complaints:", unique_edges_to_G2)

# Compare the two graphs
compare_graphs(G_complaints, G_noncomplaints)
```
